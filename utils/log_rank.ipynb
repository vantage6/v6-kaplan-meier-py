```python
import os
import base64

import pandas as pd

import pandas as pd
import numpy as np
from scipy.stats import chi2

from vantage6.algorithm.tools.mock_client import MockAlgorithmClient
```

```python
dataset_1 = {"database": "../v6-kaplan-meier-py/local/data1.csv", "db_type": "csv"}
dataset_2 = {"database": "../v6-kaplan-meier-py/local/data2.csv", "db_type": "csv"}
dataset_3 = {"database": "../v6-kaplan-meier-py/local/data3.csv", "db_type": "csv"}
org_ids = ids = [0, 1, 2]

# merge the datasets into one pandas dataframe
df = pd.concat(
    [
        pd.read_csv(dataset_1["database"]),
        pd.read_csv(dataset_2["database"]),
        pd.read_csv(dataset_3["database"]),
    ]
)

client = MockAlgorithmClient(
    datasets=[[dataset_1], [dataset_2], [dataset_3]],
    organization_ids=org_ids,
    module="v6-kaplan-meier-py",
)

organizations = client.organization.list()
org_ids = ids = [organization["id"] for organization in organizations]
```

```python
STRING_ENCODING = "utf-8"
ENV_VAR_EQUALS_REPLACEMENT = "="
def _encode(string: str):
    return (
        base64.b32encode(string.encode(STRING_ENCODING))
        .decode(STRING_ENCODING)
        .replace("=", ENV_VAR_EQUALS_REPLACEMENT)
    )
```

```python
os.environ["KAPLAN_MEIER_TYPE_NOISE"] = _encode("NONE")
os.environ["KAPLAN_MEIER_MINIMUM_ORGANIZATIONS"] = _encode("1")
```

```python
# Compute the Kaplan-Meier curve for all individual organizations
results = []
for org_id in org_ids:
    average_task = client.task.create(
        input_={
            "method": "kaplan_meier_central",
            "kwargs": {
                "time_column_name": "TIME_AT_RISK",
                "censor_column_name": "MORTALITY_FLAG",
                "organizations_to_include": [org_id]
            },
        },
        organizations=[org_ids[0]],
    )

    results_ = client.result.get(average_task.get("id"))
    results.append(pd.read_json(results_))
```

```python
def compute_log_rank_from_km(df1, df2):
    """
    Compute log rank test from two Kaplan-Meier tables

    Parameters:
    df1, df2: DataFrames containing columns:
        - TIME_AT_RISK
        - observed (number of events)
        - at_risk (number at risk)
    """
    # Combine time points and sort
    all_times = pd.concat([
        df1[['TIME_AT_RISK', 'observed', 'at_risk']].assign(group=1),
        df2[['TIME_AT_RISK', 'observed', 'at_risk']].assign(group=2)
    ]).sort_values('TIME_AT_RISK')

    # Create stats dataframe
    stats = pd.DataFrame()
    stats['time'] = all_times['TIME_AT_RISK'].unique()

    # Get observed and at_risk for each group at each time point
    for group in [1, 2]:
        group_data = all_times[all_times['group'] == group]
        stats[f'observed_{group}'] = [
            group_data[group_data['TIME_AT_RISK'] == t]['observed'].sum()
            for t in stats['time']
        ]
        stats[f'at_risk_{group}'] = [
            group_data[group_data['TIME_AT_RISK'] == t]['at_risk'].iloc[0]
            if len(group_data[group_data['TIME_AT_RISK'] == t]) > 0
            else 0
            for t in stats['time']
        ]

    # Calculate totals
    stats['total_events'] = stats['observed_1'] + stats['observed_2']
    stats['total_at_risk'] = stats['at_risk_1'] + stats['at_risk_2']

    # Calculate expected events
    stats['expected_1'] = stats['at_risk_1'] * (stats['total_events'] / stats['total_at_risk'])
    stats['expected_2'] = stats['at_risk_2'] * (stats['total_events'] / stats['total_at_risk'])

    # Calculate chi-square statistic
    O1 = stats['observed_1'].sum()
    E1 = stats['expected_1'].sum()
    O2 = stats['observed_2'].sum()
    E2 = stats['expected_2'].sum()

    chi_square = (((O1 - E1) ** 2) / E1) + (((O2 - E2) ** 2) / E2)
    p_value = 1 - chi2.cdf(chi_square, df=1)

    return chi_square, p_value, stats
```

```python
def compute_log_rank_from_km_multiple(*dfs):
    """
    Compute log rank test from N Kaplan-Meier tables

    Parameters:
    *dfs: Variable number of DataFrames, each containing columns:
        - TIME_AT_RISK
        - observed (number of events)
        - at_risk (number at risk)

    Returns:
    chi_square: The chi-square statistic
    p_value: The p-value for the test
    stats: DataFrame containing the test statistics
    """
    # Number of groups
    n_groups = len(dfs)

    if n_groups < 2:
        raise ValueError("At least two groups are required for the log-rank test")

    # Combine time points and sort
    all_times = pd.concat([
        df[['TIME_AT_RISK', 'observed', 'at_risk']].assign(group=i)
        for i, df in enumerate(dfs, 1)
    ]).sort_values('TIME_AT_RISK')

    # Create stats dataframe
    stats = pd.DataFrame()
    stats['time'] = all_times['TIME_AT_RISK'].unique()

    # Get observed and at_risk for each group at each time point
    for group in range(1, n_groups + 1):
        group_data = all_times[all_times['group'] == group]
        stats[f'observed_{group}'] = [
            group_data[group_data['TIME_AT_RISK'] == t]['observed'].sum()
            for t in stats['time']
        ]
        stats[f'at_risk_{group}'] = [
            group_data[group_data['TIME_AT_RISK'] == t]['at_risk'].iloc[0]
            if len(group_data[group_data['TIME_AT_RISK'] == t]) > 0
            else 0
            for t in stats['time']
        ]

    # Calculate totals
    stats['total_events'] = sum(stats[f'observed_{i}'] for i in range(1, n_groups + 1))
    stats['total_at_risk'] = sum(stats[f'at_risk_{i}'] for i in range(1, n_groups + 1))

    # Calculate expected events for each group
    for group in range(1, n_groups + 1):
        stats[f'expected_{group}'] = stats[f'at_risk_{group}'] * (stats['total_events'] / stats['total_at_risk'])

    # Calculate chi-square statistic
    chi_square = 0
    for group in range(1, n_groups + 1):
        O = stats[f'observed_{group}'].sum()
        E = stats[f'expected_{group}'].sum()
        if E > 0:  # Avoid division by zero
            chi_square += ((O - E) ** 2) / E

    # Degrees of freedom is (number of groups - 1)
    df = n_groups - 1
    p_value = 1 - chi2.cdf(chi_square, df=df)

    return chi_square, p_value, stats

def interpret_log_rank_results_multiple(chi_square_stat, p_value, n_groups):
    """
    Interpret the results of the multiple group log-rank test
    """
    print(f"Log-rank test results for {n_groups} groups:")
    print(f"Chi-square statistic: {chi_square_stat:.4f}")
    print(f"Degrees of freedom: {n_groups - 1}")
    print(f"p-value: {p_value:.4f}\n")

    print("Conclusion:")
    if p_value < 0.05:
        print(f"The log-rank test shows a statistically significant difference between the survival curves (p = {p_value:.4f}). "
              f"This suggests that there are significant differences in survival experiences among the {n_groups} groups.")
    else:
        print(f"The log-rank test does not show a statistically significant difference between the survival curves (p = {p_value:.4f}). "
              f"This suggests that there is not enough evidence to conclude that the survival experiences differ among the {n_groups} groups.")

# Example usage:
# chi_square_stat, p_value, stats = compute_log_rank_from_km_multiple(df1, df2, df3, ...)
# interpret_log_rank_results_multiple(chi_square_stat, p_value, n_groups=len([df1, df2, df3, ...]))
```
